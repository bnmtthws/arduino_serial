```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
testfile = 'arduino_sixPin-20141002-083943.csv'
```

####read data and convert index to time

```python
data = pd.read_csv(testfile, delimiter=',',names=['time','h1','h2','h3','t1','t2','t3'])
data.index = pd.to_datetime(data['time'],format='%H%M%S')
```

####plot raw data - note the spurious 

```python
data[['h1','h2','h3','t2']].plot()
plt.ylim(0,100)
```

####plot as rolling median

```python
pd.rolling_median(data[['h1','h2','h3','t2']],50).plot()
```
